## Create WLIA Workshop Environment


```python
# Import the ArcGIS API for Python
from arcgis.gis import GIS

# Connect to the current AGO organization. In ArcGIS Online notebooks, "home" will use the current credentials.
gis = GIS("home")
```

```python
# Connect to the RM organization to get access to the source web map and items (Use AGO Assistant as another option to get around this)
source_gis = GIS("https://r-m.maps.arcgis.com", "", "")
```

### Input arguments. In this case:

```python
# source_webmap_id is the ItemID of the AGO web map to be cloned over for the workshop.
source_webmap_id = "a62634ec18714f0b90af09d1dec220aa"

# A folder that we will create for organizational purposes
target_folder = "WLIA HFL Workshop"
```

### Create the folder

```python
# Create a WLIA folder for  all the content
gis.content.folders.create(target_folder)
```

### Clone the Web Map

```python
# Clone the web map (this also clones all the dependent items, in this case a feature layer and a view layer))
cloned_items = gis.content.clone_items(
    items=[source_gis.content.get(source_webmap_id)],
    folder=target_folder
)
```

### Print the items so that we can use them for the workshop

```python
for item in cloned_items:
    print(f"Cloned item title: {item.title}")
    print(f"Cloned item ID: {item.id}")
    print(f"Cloned item URL: {item.url if hasattr(item, 'url') else 'N/A'}")
```
